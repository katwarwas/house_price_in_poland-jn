### *Importing important libraries*

```python
import matplotlib
# matplotlib.use('TkAgg')
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### *Data loading*

```python
df = pd.read_csv('Apartments_in_poland.csv')
max_price_index = df['price per m2'].idxmax()
df.drop(max_price_index, inplace=True)
df = df[df['area'] <= 500]
df.head()
```

### *Poland*

#### *Area density chart*

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.kdeplot(df['area'], color='g', fill=True, alpha=0.3)
plt.axvline(df['area'].median(), color='black')
plt.axvline(df['area'].mean(), color='red')
plt.title('Apartment area distribution')
plt.xlabel('Area')
plt.ylabel('Density')
plt.show()
df['area'].median()
```

#### *Price density chart*

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['price'], color='g', fill=True, alpha=0.3)
plt.axvline(df['price'].median(), color='black')
plt.axvline(df['price'].mean(), color='red')
plt.title('Apartment price distribution')
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()
```

#### *Price per m^2 density chart*

```python
filtered_data = df[df['price per m2'] <= 200000]

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.kdeplot(filtered_data['price per m2'], color='g', fill=True, alpha=0.3)
plt.axvline(filtered_data['price per m2'].median(), color='black')
plt.axvline(filtered_data['price per m2'].mean(), color='red')
plt.title('Apartment Price per m2 Distribution')
plt.xlabel('Price per m2')
plt.ylabel('Density')
plt.show()
```

#### *Room counts*

```python
room_counts = df['number of rooms'].value_counts()

room_counts_combined = room_counts.copy()
room_counts_combined.loc[room_counts_combined.index > 5] = room_counts_combined[room_counts_combined.index > 5].sum()
room_counts_combined.loc[5] = room_counts_combined[room_counts_combined.index > 5].sum()
room_counts_combined = room_counts_combined[room_counts_combined.index <= 5]

labels = room_counts_combined.index.astype(str).tolist()
labels[-1] = '5+'

plt.pie(room_counts_combined, labels=labels, autopct='%1.1f%%')
plt.title('Apartments - Room Counts')
plt.axis('equal')
plt.show()
```

#### *Distribution of Room Counts by Voivodeship*

```python
grouped_data = df.groupby(['voivodeship', 'number of rooms']).size().unstack()

grouped_data.loc[:, '5+'] = grouped_data.iloc[:, 5:].sum(axis=1)
grouped_data = grouped_data.iloc[:, :5]
grouped_data = grouped_data.drop(columns=[i for i in range(5, grouped_data.shape[1])])
grouped_data = grouped_data.rename(columns={grouped_data.columns[-1]: '5+'})

plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', stacked=True)
plt.title('Apartments - Distribution of Room Counts by Voivodeship')
plt.xlabel('Voivodeship')
plt.ylabel('Count of Rooms')
plt.legend(title='Number of Rooms',loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

#### *Percentage Distribution of Room Counts by Voivodeship*

```python
grouped_data = df.groupby('voivodeship')['number of rooms'].value_counts(normalize=True).unstack() * 100

grouped_data.loc[:, '5+'] = grouped_data.iloc[:, 5:].sum(axis=1)
grouped_data = grouped_data.iloc[:, :5]
grouped_data = grouped_data.drop(columns=[i for i in range(5, grouped_data.shape[1])])
grouped_data = grouped_data.rename(columns={grouped_data.columns[-1]: '5+'})

# plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', stacked=True)
plt.title('Apartments - Percentage Distribution of Room Counts by Voivodeship')
plt.xlabel('Voivodeship')
plt.ylabel('Percentage')
plt.legend(title='Number of Rooms',loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

#### *Percentage Distribution of Area*

```python
bins = [0, 30, 50, 80, 120, float('inf')]
labels = ['<30', '30-50', '50-80','80-120', '>120']

# Tworzenie kolumny z przedziałami powierzchni
df['area_category'] = pd.cut(df['area'], bins=bins, labels=labels)

# Obliczanie procentowego rozkładu powierzchni dla każdego województwa
grouped_data = df['area_category'].value_counts(normalize=True) * 100

# Tworzenie wykresu kołowego
plt.pie(grouped_data, labels=grouped_data.index, autopct='%1.1f%%')
plt.title('Apartments - Percentage Distribution of Area')
plt.axis('equal')
plt.show()
```

#### *Percentage Distribution of Area by Voivodeship*
